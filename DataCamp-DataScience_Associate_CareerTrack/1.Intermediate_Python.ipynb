### *Log-log Plots*

A log-log plot is a type of graph that uses logarithmic scales on both the horizontal (x-axis) and vertical (y-axis) axes. Instead of plotting the raw values, it plots the logarithm of the values. This means each axis is scaled logarithmically, so equal distances represent equal ratios rather than equal differences.

#### When to Use Log-Log Plots?

Data Spans Multiple Orders of Magnitude:
If your data ranges from very small to very large values, a log-log plot can help visualize all data points effectively.


Suspected Power-Law Relationship:
    
If you think the relationship between your variables follows a power law, a log-log plot will linearize the data, allowing you to use linear regression techniques.

Analyzing Fractal or Self-Similar Data: 

Log-log plots are useful in fields like physics, biology, and economics, where self-similar patterns occur across different scales.

```python
k = 1
n = 2

x = np.linspace(1, 1000, 1000)

# Compute y values using the power-law formula
y = k * x**n
```

```python
plt.figure(figsize=(10, 5))

# Plot data on a log-log scale
plt.loglog(x, y, label='y = x^2')


plt.xlabel('X Values (log scale)')
plt.ylabel('Y Values (log scale)')
plt.title('Log-Log Plot of a Power-Law Relationship')

plt.grid(True, which="both", ls="--")
plt.legend()

plt.show()
```

```python
np_height = np.array([1.73 , 1.88 , 1.99 , 2.05])
np_weight = np.array([65.4 , 59.2 , 63.6 , 88.1])

meas = np.array([np_height , np_weight])

# Loop through the 2D array with an index
for i, val in enumerate(np.nditer(meas)):
    
    # Determine if it's height or weight based on the index
    if i < len(np_height):
        print(f"{type(val)}  ---  This is height: {val}")
    else:
        print(f"{type(val)}  ---  This is weight: {val}")
```

```python
# simulating 1000 dice throws

# array to store results
results = []

# Simulating the process 1000 times
for _ in range(1000):
    step = 50
    dice = np.random.randint(1, 7)

    if dice <= 2:
        step = step - 1
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1, 7)

    # Append dice and step to results
    results.append((dice, step))

# Print out the first few results to check
for dice, step in results[:5]:
    print(f"Dice: {dice}, Step: {step}")


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Array to store results
results = []

# Simulating the process 100 times
for _ in range(100):
    step = 50
    dice = np.random.randint(1, 7)

    if dice <= 2:
        step = step - 1
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1, 7)

    # Append dice and step to results
    results.append((dice, step))

# Convert results to a NumPy array for easier indexing
results_array = np.array(results)

# Extract dice and step values
dice_results = results_array[:, 0]
step_results = results_array[:, 1]

# Create a line plot for steps over the simulations
plt.figure(figsize=(12, 6))
plt.plot(step_results, label='Step', color='blue', linewidth=2)
plt.title('Step Value Over 1000 Simulations')
plt.xlabel('Simulation Number')
plt.ylabel('Step Value')
plt.axhline(y=50, color='red', linestyle='--', label='Starting Point (Step=50)')
plt.legend()
plt.grid()
plt.show()
```

## *RANDOM WALK*

```python
np.random.seed(123)
outcomes = []

for x in range(10):
    coin = np.random.randint(0,2)
    if coin == 0:
        outcomes.append("heads")
        
    else:
        outcomes.append("tails")
        
print(outcomes)
```

```python
tails = [0]

for x in range(10):
    coin = np.random.randint(0,2)
    tails.append(tails[x] + coin)
    
print(tails)
```

```python
random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)

    if dice <= 2:
        # use max to make sure step can't go below 0
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    random_walk.append(step)

print(random_walk)
```

```python
plt.figure(figsize=(10,5))
plt.plot(random_walk)
plt.show()
```

```python
# initialize and populate all_walks
all_walks = []
for i in range(5) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
# plt.plot(np_aw)
# plt.show()

# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np_aw.T

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

## *DISTRIBUTION*

once the distribution is known it might be possible to start calculating chance of events occuring or not occuring

```python
final_tails = []

for x in range(500000):
    tails = [0]
    
    for x in range(10):
        coin = np.random.randint(0,2)
        tails.append(tails[x] + coin)
        
    final_tails.append(tails[-1])
    
#print(final_tails)   
```

```python
plt.hist(final_tails, bins=10)
plt.show()
```

```python
# Simulate random walk 500 times
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```

```python
print( f' {np.mean(ends >= 60)} % CHANCE')
```
