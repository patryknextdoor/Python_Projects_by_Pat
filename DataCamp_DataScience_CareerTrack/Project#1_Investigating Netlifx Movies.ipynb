### Project desctiption

Perform exploratory data analysis on the netflix_data.csv data to understand more about movies from the 1990s decade.

What was the most frequent *movie duration in the 1990s*? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).

A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

```python
import pandas as pd
import matplotlib.pyplot as plt


netflix_df = pd.read_csv("netflix_data.csv")

netflix_df.head(3)
netflix_df.tail(3)
```

```python
# Filter the data for movies released in the 1990s
```

```python
netflix_df = netflix_df[netflix_df["type"] == "Movie"]

new_df = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999)]
new_df.head()
```

```python
# Find the most frequent movie duration
```

```python
duration = new_df['duration']
```

```python
plt.hist(duration, bins=20 )
plt.ylabel("Frequency")
plt.xlabel("Duration in minutes")
plt.show()
```

```python
#Count the number of short action movies from the 1990s
```

```python
action_df = new_df[new_df['genre'] == 'Action']
action_df.head(3)
```

```python
short_action_movies = action_df[action_df["duration"] < 90]
count = short_action_movies.shape[0]
print(type(count), count)
```

```python
# or

short_movie_count = 0

for label , row in action_df.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1
        
print(type(short_movie_count) , short_movie_count)
```
